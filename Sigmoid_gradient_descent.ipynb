# Plotting Sigmoid function in 2D

```
import numpy as np
import matplotlib.pyplot as plt
```

$S_{w,b}(x) = \frac{1}{1+e^{-(wx+b)}}$

```
def sigmoid(x,w,b):
  return (1/(1+np.exp(-(w*x+b))))
```

```
sigmoid(0,0,0)
```

```
sigmoid(1,0.5,0)
```

```
w = 1.3   #@param {type: "slider", min: -3, max: 3,step: 0.1}
b = 0.1   #@param {type: "slider", min: -3, max: 3,step: 0.1}
x = np.linspace(-10,10,100)
Y = sigmoid(x,w,b)  # This is doing python broadcasting where it is applying the function to individual value of element in the array x.
```

```
plt.plot(x,Y)
plt.show()
```

# Sigmoid in 3D plot

$S_{w,b}(x) = \frac{1}{1+{e^-(w_1x_1+w_2x_2+b)}}$

```
def sigmoid_3d(x1,x2,w1,w2,b):
  return (1/(1+np.exp(-(w1*x1+w2*x2+b))))
```

```
sigmoid_3d(1,0,0.5,0,0)
```

```
# library to plot a 3D plot
from mpl_toolkits import mplot3d
```

```
x1 = np.linspace(-10,10,100)
x2 = np.linspace(-10,10,80)

xx1,xx2 = np.meshgrid(x1,x2)

print(x1.shape,x2.shape,xx1.shape,xx2.shape)
```

```
w1 = 0.5
w2 = 0.5
b = 0
Y = sigmoid_3d(xx1,xx2,w1,w2,b)
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(xx1,xx2,Y,cmap='viridis')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')

ax.view_init(30,270)
fig
```
